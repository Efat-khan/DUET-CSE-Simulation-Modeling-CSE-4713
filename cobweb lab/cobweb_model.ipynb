```python
import numpy as np
import matplotlib.pyplot as plt

# Define the demand and supply functions
def demand(price):
    return 12.4 - 1.2 * price

def supply(price_previous):
    return 8.0 - 0.6 * price_previous

# Calculate the equilibrium price and quantity
def find_equilibrium():
    # Set demand equal to supply and solve for P
    # 12.4 - 1.2*P = 8.0 - 0.6*P
    # Rearranging to find P:
    P_star = (12.4 - 8.0) / (1.2 - 0.6)
    Q_star = demand(P_star)  # Use the demand equation to find Q
    return P_star, Q_star

# Find the equilibrium price and quantity
P_star, Q_star = find_equilibrium()

# Number of periods (iterations)
periods = 10

# Initialize lists to store price and quantity values for each period
P0 = 1.0  # Initial price
prices = [P0]
quantities = [demand(P0)]

# Compute price and quantity for each period
for t in range(1, periods):
    price_previous = prices[t - 1]
    quantity_demanded = demand(price_previous)
    prices.append(supply(price_previous))  # Next price from supply
    quantities.append(quantity_demanded)   # Quantity from demand

# Parameters for Stable Market (a)
P0_stable = 1.0
a_stable = 12.4
b_stable = 1.2
c_stable = 1.0
d_stable = 0.9

# Parameters for Unstable Market (b)
P0_unstable = 5.0
a_unstable = 10.0
b_unstable = 0.9
c_unstable = -2.4
d_unstable = 1.2

# Time Periods
time_periods = 6

# Generate Stable Market Data
prices_stable = [P0_stable]
for t in range(1, time_periods + 1):
    P_next = a_stable - b_stable * prices_stable[-1] + c_stable * t - d_stable * prices_stable[-1]
    prices_stable.append(P_next)

# Generate Unstable Market Data
prices_unstable = [P0_unstable]
for t in range(1, time_periods + 1):
    P_next = a_unstable - b_unstable * prices_unstable[-1] + c_unstable * t - d_unstable * prices_unstable[-1]
    prices_unstable.append(P_next)

# Plotting
fig, ax = plt.subplots(2, 1, figsize=(10, 12))

# Subplot 1: Stable and Unstable Market
ax[0].plot(range(0, time_periods + 1), prices_stable, marker='o', color='green', linestyle='-', label='Stable Market')
ax[0].plot(range(0, time_periods + 1), prices_unstable, marker='o', color='red', linestyle='--', label='Unstable Market')
ax[0].set_title('Stable vs Unstable Market')
ax[0].set_xlabel('Time Period')
ax[0].set_ylabel('Price')
ax[0].legend()
ax[0].grid(True)

# Subplot 2: Cobweb Graph
price_range = np.linspace(0, 10, 100)
demand_curve = demand(price_range)
supply_curve = [supply(p) for p in price_range]
ax[1].plot(demand_curve, price_range, label='Demand', color='blue', linewidth=2)
ax[1].plot(supply_curve, price_range, label='Supply', color='orange', linewidth=2)

# Add Cobweb Path
cobweb_prices = [P0]
cobweb_quantities = [demand(P0)]
for t in range(1, periods):
    cobweb_prices.append(prices[t])
    cobweb_quantities.append(demand(prices[t]))
    ax[1].plot([cobweb_quantities[-2], cobweb_quantities[-1]], [cobweb_prices[-2], cobweb_prices[-2]], 'r--')  # Horizontal
    ax[1].plot([cobweb_quantities[-1], cobweb_quantities[-1]], [cobweb_prices[-2], cobweb_prices[-1]], 'r--')  # Vertical

ax[1].plot(Q_star, P_star, 'go', label=f'Equilibrium Point ({Q_star:.2f}, {P_star:.2f})')
ax[1].set_title('Cobweb Model')
ax[1].set_xlabel('Quantity')
ax[1].set_ylabel('Price')
ax[1].legend()
ax[1].grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```
